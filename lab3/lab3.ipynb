```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier # Multi-layer Perceptron Classifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# perceptron is a type of artificial neural network that consists of a single layer of output nodes
```

```python
# import the data
data = pd.read_csv("diabetes.csv")
data
```

```python
# independent and dependent
# what are the independent and dependent variables?
# independent variables are the features used for prediction
# dependent variable is the target variable we want to predict

x = data.iloc[:,:8] # [:,:8] means all rows and columns 0 to 7
y = data.iloc[:,8] # [:,8] means all rows and column 8

```

```python
x
```

```python
y
```

02) Split the data into training and testing sets

• Which function will you use to split the dataset into training and testing sets?

• How will you decide the proportion for training and testing data?

• Why might it be important to set a fixed random seed when splitting data?

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
print(data["Outcome"].unique())
```

03) Create the MLP Classifier model object

• Which scikit-learn class is used to create an MLP model for classification?

• How can you define the architecture of the hidden layers and the activation
function?

• Which parameters control the optimization process and the number of training
iterations?

```python
clf = MLPClassifier(alpha=0.01,hidden_layer_sizes=(5,3), random_state=1, max_iter=1000)

# random state is used to ensure reproducibility of results
```

```python
clf
```

04) Train the model with training data

• Which method will you use to train the model?

• What arguments will you pass to this method?

```python
clf.fit(x_train,y_train)
```

05) Check the accuracy of the testing data

• Which method can directly return the accuracy of the model on the test dataset?

• How can you calculate accuracy using the metrics module in scikit-learn?

• What could cause differences between training accuracy and testing accuracy?

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
