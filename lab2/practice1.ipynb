```python
import pandas as pd 
import numpy as np
```

```python
data1 = {
    "Name": ["Alice", "Bob", np.nan, "David"],
    "Age": [25, np.nan, 22, 28],
    "City": ["Colombo", "Kandy", "Galle", np.nan],
}
df1 = pd.DataFrame(data1)
df1
```

```python
df1.isnull()
```

```python
df1_filled=df1.fillna({
    "Name" : "Unknown",
    "Age" : df1["Age"].mean(),
    "City" : "Unknown"
})
df1_filled
```

```python
df1.dropna()
```

```python
data3 = {
'ID': ['101', '102', '103'],
'JoinDate': ['2022-01-10', '2022-02-15',
'2022-03-01'],
'Salary': ['50000', '60000', '70000']
}

df3 = pd.DataFrame(data3)
df3
```

```python
df3.dtypes
```

```python
df3["ID"] = df3["ID"].astype(int)
df3["JoinDate"] = pd.to_datetime(df3["JoinDate"])
df3["Salary"] = df3["Salary"].astype(float)

df3.dtypes
```

```python
data4 = {"Product": [" Apple ", "BaNaNa", " Cherry", "DaTE "]}
df4 = pd.DataFrame(data4)
print(df4)
```

```python
df4["Product"] = df4["Product"].str.strip()
print(df4)
```

```python
df4["Product"] = df4["Product"].str.lower()
print(df4)
```

```python
data5 = {
'Score': [55, 60, 62, 90, 95, 200, 58, 59, 210]
}
df5 = pd.DataFrame(data5)
df5
```

```python
df5['Score']>100
```

```python
df5[df5['Score']>100]
```

```python
mean_score = df5[df5["Score"]<=100]["Score"].mean()
mean_score
```

```python
df5["Score"] = df5["Score"].apply(lambda x:mean_score if x > 100 else x)
df5
```
